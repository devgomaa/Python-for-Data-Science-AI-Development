## One Sampled Z-test




x bar is the Sample mean

mu is the population mean

sigma is the population standard deviation

n is the sample size

```python
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
```

```python
# lets check the head of the dataset
pd.set_option('max_columns', 1000)
data.head()
```

## We are testing whether the mean of house prices is 180000 or not

Null Hypothesis :- The mean house prices is 180000

Alternate Hypothesis :- The mean house prices is not 180000

```python
from statsmodels.stats import weightstats as stests

#from scipy import stats

ztest ,pval = stests.ztest(x1 = data['SalePrice'], x2=None, value=180000)
print("P Value :",float(pval))
print("z-test value :",float(ztest))

if pval<0.05:
    print(" We reject the null hypothesis")
else:
    print("We fail to reject the null hypothesis")
```

## Two Sampled z-test

## To Check if there is an association between the prices of the First floor per square feet and Second-floor per square feet houses. 


Null Hypothesis:- The means of the first floor and the second floor per feet square houses are equal.

Alternate Hypothesis:- The means of the first floor and the second floor per feet square houses are not equal.

```python
zstats, pval = stests.ztest(data['1stFlrSF'], data['2ndFlrSF'], value = 0)
print("\np-value",pval)
print("z-score ",zstats)

if pval <0.05:
    print("we reject the null hypothesis")
else:
    print("we fail to reject the null hypothesis")
```
